```python
#MTA Subway Major incidents
from google.colab import files
import pandas as pd

df = pd.read_csv('/content/drive/My Drive/MTA_Subway_Major_Incidents__Beginning_2020.csv')
```

```python
df.head()
```

- Plot MTA Subway Major incidents

```python
grouped_data = df.groupby("category")['count'].sum()
```

```python
import matplotlib.pyplot as plt

grouped_data = grouped_data.sort_values(ascending=False)

# Create the bar plot
ax = grouped_data.plot(kind="bar")

# Customize the x-axis labels to prevent overlapping
plt.xticks(rotation=40, fontsize=10, fontstyle='italic', ha='right')

#labels and title
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("MTA Subway Major Incidents")

#Plot
plt.show()
```
